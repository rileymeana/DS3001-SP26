Q2. Being able to make basic plots to visualize sets of points is really helpful, particularly with data analysis. The pyplot plots are built up slowly by defining elements of the plot, and then using plt.show() to create the final plot. This question gives you some practice doing that iterative building process.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a grid of 50 points ranging from a small positive number to 1
# Starting slightly above 0 to avoid log(0) which is undefined.
x = np.linspace(0.001, 1, 50)

# Create numpy array y for natural logarithm
y = np.log(x)

# Create numpy array z for exponential function
z = np.exp(x)

# Create a single plot for both functions
plt.figure(figsize=(10, 6))

# Plot Natural Logarithm
plt.plot(x, y, color='blue', label='Natural Log (ln(x))')

# Plot Exponential Function
plt.plot(x, z, color='red', label='Exponential (exp(x))')

# Add title and labels
plt.title('Natural Logarithm and Exponential Functions')
plt.xlabel('x')
plt.ylabel('Function Value')
plt.grid(True)
plt.legend(loc='upper left') # Legend in the upper left to avoid overlap

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a grid of 100 equally spaced points ranging from -6.5 to 6.5
x_new = np.linspace(-6.5, 6.5, 100)

# Compute sine and cosine values
y_sin = np.sin(x_new)
y_cos = np.cos(x_new)

# Create the plot
plt.figure(figsize=(10, 6))

# Plot sine function as a line graph
plt.plot(x_new, y_sin, label='Sine Function (sin(x))', color='green')

# Plot cosine function as a line graph
plt.plot(x_new, y_cos, label='Cosine Function (cos(x))', color='purple', linestyle='--')

# Add labels, title, and legend
plt.xlabel('x')
plt.ylabel('Function Value')
plt.title('Sine and Cosine Functions')
plt.legend(loc='lower left')

# Add grid for better readability
plt.grid(True)

# Display the plot
plt.show()
```
